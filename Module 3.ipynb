# Assignment 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D

# Look pretty...
# matplotlib.style.use('ggplot')
plt.style.use('ggplot')
```

```python
data = pd.read_csv('wheat.data')
data.head()
```

```python
# slice the dataframe: area and perimeter features
s1 = data[['area','perimeter']]
s1.head()
```

```python
# slice the dataframe: groove and asymmetry
s2 = data[['groove','asymmetry']]
s2.head()
```

```python
plt.figure()
s1.plot.hist(alpha=0.7)
plt.show()
#perimeter is more normal
```

```python
plt.figure()
s2.plot.hist(alpha=0.7)
plt.show()

# asymmetry has more variance
```

# Assignment 2

```python
plt.figure()
s1.plot.scatter(x='area',y='perimeter')
plt.show()
```

```python
plt.figure()
s2.plot.scatter(x='groove',y='asymmetry')
plt.show()
```

```python
s3 = data[['compactness','width']]
plt.figure()
s3.plot.scatter(x='compactness',y='width')
plt.show()
```

# Assignment 3

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Asymmetry')

ax.scatter(data.area, data.perimeter, data.asymmetry, c='r', marker='.')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Width')
ax.set_ylabel('Groove')
ax.set_zlabel('Length')

ax.scatter(data.width, data.groove, data.length, c='g', marker='.')
plt.show()
```
